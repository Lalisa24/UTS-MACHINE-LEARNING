```python
# Assign default values for each data type
defaultInt = 0
defaultString = 'NaN'
defaultFloat = 0.0

# Create lists by data tpe
intFeatures = ['Age']
stringFeatures = ['Gender', 'Country', 'self_employed', 'family_history', 'treatment', 'work_interfere',
                 'no_employees', 'remote_work', 'tech_company', 'anonymity', 'leave', 'mental_health_consequence',
                 'phys_health_consequence', 'coworkers', 'supervisor', 'mental_health_interview', 'phys_health_interview',
                 'mental_vs_physical', 'obs_consequence', 'benefits', 'care_options', 'wellness_program',
                 'seek_help']
floatFeatures = []

# Clean the NaN's
for feature in df:
    if feature in intFeatures:
        df[feature] = df[feature].fillna(defaultInt)
    elif feature in stringFeatures:
        df[feature] = df[feature].fillna(defaultString)
    elif feature in floatFeatures:
        df[feature] = df[feature].fillna(defaultFloat)
    else:
        print('Error: Feature %s not recognized.' % feature)
df.head(5)
```

Membersihkan data gender

```python
#clean 'Gender'
#Slower case all columm's elements
gender = df['Gender'].str.lower()
#print(gender)

#Select unique elements
gender = df['Gender'].unique()

#Made gender groups
male_str = ["male", "m", "male-ish", "maile", "mal", "male (cis)", "make", "male ", "man","msle", "mail", "malr","cis man", "Cis Male", "cis male"]
other_str = ["trans-female", "something kinda male?", "queer/she/they", "non-binary","nah", "all", "enby", "fluid", "genderqueer", "androgyne", "agender", "male leaning androgynous", "guy (-ish) ^_^", "trans woman", "neuter", "female (trans)", "queer", "ostensibly male, unsure what that really means"]
female_str = ["cis female", "f", "female", "woman",  "femake", "female ","cis-female/femme", "female (cis)", "femail"]

for (row, col) in df.iterrows():
   if str.lower(col.Gender) in male_str:
        df['Gender'].replace(to_replace=col.Gender, value='male', inplace=True)

   if str.lower(col.Gender) in female_str:
        df['Gender'].replace(to_replace=col.Gender, value='female', inplace=True)

   if str.lower(col.Gender) in other_str:
        df['Gender'].replace(to_replace=col.Gender, value='other', inplace=True)

#Get rid of bullshit
stk_list = ['A little about you', 'p']
df = df[~df['Gender'].isin(stk_list)]

print(df['Gender'].unique())
```

membuat chisquare

```python
from scipy.stats import chi2_contingency
import math

cat_cols = ['self_employed', 'family_history', 'treatment',
            'tech_company', 'benefits', 'care_options','mental_health_interview']

# Create cross tables and heat maps between treatment and other categorical columns
significant_cols = []
for col in cat_cols:
    if col == 'treatment':
        continue

    # Create a cross table
    crosstab = pd.crosstab(df['treatment'], df[col])
    # Apply Chi-square test
    chi2, p, dof, expected = chi2_contingency(crosstab)

    # Select columns with p-value less than 0.05
    if p < 0.05:
        significant_cols.append(col)
        print(f'Chi-square test between treatment and {col}:')
        print(f'Chi2 Value: {chi2}, p Value: {p}\n')

# Count the selected columns
num_plots = len(significant_cols)
num_cols = 4  # Number of plots per row
num_rows = math.ceil(num_plots / num_cols)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, num_rows * 5))
axes = axes.flatten()
for i, col in enumerate(significant_cols):
    # Create a cross table
    crosstab = pd.crosstab(df['treatment'], df[col])

    # Visualize the cross table as a heat map
    sns.heatmap(crosstab, annot=True, cmap='coolwarm', fmt='g', ax=axes[i])
    axes[i].set_title(f'{col} and treatment')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('treatment')

# Hide empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

Dari heatmap pertama, terlihat bahwa individu yang memiliki riwayat keluarga dengan masalah kesehatan mental (family_history = 0) lebih banyak yang tidak mencari perawatan (112 orang) dibandingkan yang mencari perawatan (86 orang). Supaya lebih stabil, kita tambahkan Random Forest. Caranya, kita bikin banyak Decision Tree, terus hasilnya digabung biar lebih akurat. Jadi, kalau satu pohon salah, pohon lain bisa bantu benerin. Tapi ternyata, Random Forest ini masih punya kelemahan: prosesnya berat dan kurang efisien, apalagi kalau datanya banyak.

Setelah menggunakan Random Forest, pendekatan ensemble learning yang menggabungkan banyak pohon keputusan secara paralel, langkah selanjutnya dalam meningkatkan performa model adalah Boosting. Berbeda dari Random Forest yang membangun model secara independen dan menggabungkan hasilnya melalui voting atau rata-rata, Boosting bekerja secara berurutan, di mana setiap model baru berusaha memperbaiki kesalahan dari model sebelumnya.

Dalam Boosting, setiap iterasi memberikan bobot lebih besar pada data yang sulit diprediksi. Ini memungkinkan model untuk lebih fokus pada kesalahan dan meningkatkan akurasi. Salah satu pengembangan dari Boosting adalah Gradient Boosting, yang menggunakan pendekatan optimasi berbasis gradient descent. Alih-alih hanya menyesuaikan bobot sampel, Gradient Boosting membangun model baru untuk memperbaiki kesalahan model sebelumnya dengan menghitung selisih antara prediksi dan nilai sebenarnya (residual error).

Namun, meskipun Gradient Boosting sudah cukup kuat, metode ini masih memiliki beberapa kelemahan, seperti waktu komputasi yang lama dan risiko overfitting. Oleh karena itu, muncullah XGBoost (Extreme Gradient Boosting) sebagai solusi yang lebih efisien. XGBoost mengoptimalkan Gradient Boosting dengan beberapa teknik seperti regularisasi, parallel processing, dan pruning untuk membuat model lebih cepat, lebih efisien dalam penggunaan memori, dan lebih tahan terhadap overfitting.

Nah, di sinilah XGBoost masuk. Model ini nggak cuma bikin banyak pohon kayak Random Forest, tapi juga belajar dari kesalahan sebelumnya. Jadi, setiap kali model bikin prediksi yang salah, pohon berikutnya bakal nyoba memperbaikinya. Ini bikin modelnya jauh lebih akurat, cepat, dan nggak gampang overfitting.
